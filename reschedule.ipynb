# Stack all works for each grid

```python
local_works = {}
for grid in project.grids:
    works = [work for work in project.schedule if work.grid.location == grid.location]
    local_works[grid.location] = sorted(works, key=lambda x:x.day, reverse=False)
```

```python
local_works
```

# Update schedule

```python
print(project.duration)
```

```python
def __check_productivity(activity_stack, next_activity):
    '''
    If the number of activities existing in the stack is smaller than the activity's productivity, return True.
    '''

    if activity_stack[next_activity.code] < project.activities[next_activity.code].productivity:
        return True
    else:
        return False
```

```python
updated_schedule = []

day = 0
while day <= project.duration:
    ## Initialize activity stack.
    activity_stack = {activity_code: 0 for activity_code in project.activities.keys()}

    ## Try to assign a work on each grid.
    ## TODO: Start at a grid with the longest workdays and move to the nearest grid.
    for grid in project.sorted_grids:
        remaining_works = local_works[grid.location]
        if not remaining_works:
            continue

        ## Get the first remaining work on the grid.
        next_work = remaining_works[0]
        next_activity = next_work.activity

        ## If productivity on the day is full, remain the grid blank and move to the next.
        if not __check_productivity(activity_stack=activity_stack, next_activity=next_activity):
            continue

        next_work = remaining_works.pop(0)
        work = Work(grid=grid, day=day, activity=next_work.activity)
        updated_schedule.append(work)
        activity_stack[work.activity.code] += 1
    
    day += 1
```

```python
project.schedule = updated_schedule
schedule_udated = project.schedule2df()
```
